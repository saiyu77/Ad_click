```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
ad = pd.read_csv('adclick.csv')
```

```python
ad.head()
```

```python
ad.info()
```

```python
le = LabelEncoder()
ad['gender_n'] = le.fit_transform(ad['gender'])
ad['device_type_n'] = le.fit_transform(ad['device_type'])
ad['ad_position_n'] = le.fit_transform(ad['ad_position'])
ad['browsing_history_n'] = le.fit_transform(ad['browsing_history'])
ad['time_of_day_n'] = le.fit_transform(ad['time_of_day'])
```

```python
ad.head()
```

```python
ad.info()
```

```python
ad = ad.drop(['full_name','gender','device_type','ad_position','browsing_history','time_of_day'],axis=1)
```

```python
ad.head()
```

```python
fea_cols = ['age','gender_n','device_type_n','ad_position_n','browsing_history_n','time_of_day_n']
X = ad[fea_cols]
Y = ad.click
```

```python
X.head()
```

```python
Y.head()
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=1)
```

```python
clo = DecisionTreeClassifier()
clo = clo.fit(X_train,Y_train)
Y_pred = clo.predict(X_test)
```

```python
print("Accuracy:",metrics.accuracy_score(Y_test, Y_pred))
```

```python
plt.figure(figsize=(150,100))
plot_tree(clo,feature_names=X.columns,class_names=['1','0'],filled=True)
plt.show()
```
